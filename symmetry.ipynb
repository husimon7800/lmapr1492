```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from IPython.display import FileLink
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.core.operations import SymmOp
from jupyter_jsmol import JsmolView
from jupyter_jsmol.pymatgen import quick_view


key = "OeM7QMOQddQoMGfFB4h"
material = "mp-866132"
material_name = "Na₂TlSb"
with MPRester(key) as m:
    struct = m.get_structure_by_material_id(material,conventional_unit_cell=True)
    symOp = SpacegroupAnalyzer(struct).get_symmetry_operations()
    sit = struct.sites
    w = CifWriter(struct)
    w.write_file(material+".cif") 
    print("Clique sur le lien pour télécharger mon fichier .CIF")
    print("Attention, ce fichier .CIF n'est pas le fichier .CIF “symmetrized”")
    display(FileLink(material+".cif"))
print(struct)
```

# Symétrie de Rotation d'ordre 2

```python

print(" Symétrie de Rotation d'ordre 2  : ")
op = SymmOp.from_xyz_string("-x,y,-z")
pos_init = struct.sites[1].frac_coords  
resultat = op.operate(pos_init)
print ("Position initiale de l'atome de Na ="+ str(pos_init))
print("=====================================================")
print ("Matrice de rotation")
print(op.rotation_matrix)
print("=====================================================")
print("Vecteur de translation")
print(op.translation_vector)
print("=====================================================")

print ("Position après application de la symétrie"+ str(resultat))

```

# Symétrie de Réflexion

```python
print("Symétrie de Réflexion : ")
Op = SymmOp.from_xyz_string('-y,-x,z')
pos= struct.sites[11].frac_coords 
result = Op.operate(pos)
print("Position initiale de l'atome de Tl" + str(pos))
print("=====================================================")
print ("Matrice de rotation")
print(Op.rotation_matrix)
print("=====================================================")
print("Vecteur de translation")
print(Op.translation_vector)
print("=====================================================")
print("Position après application de la symétrie" + str(result))
```
